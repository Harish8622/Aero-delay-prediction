# Evaluation Notebook

This notebook runs a minimal agent flow using `build_agent_app()`.


```python
import os
import sys
from pathlib import Path

cwd = Path.cwd()
root = cwd.parent if cwd.name == "notebooks" else cwd
os.chdir(root)
sys.path.append(str(root))

from agent.customer_agent import build_agent_app
from src.agent_core import agent_node_prompt
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
app = build_agent_app()
conversation_state = {"messages": [SystemMessage(content=agent_node_prompt)]}

conversation_state["messages"].append(HumanMessage(content="hi"))
result_state = app.invoke(conversation_state)
response = result_state["messages"][-1]
response.content
```

```python
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.test_case import LLMTestCase
from deepeval import evaluate

user_message = "hi"
response_text = response.content

test_case = LLMTestCase(input=user_message, actual_output=response_text)
metric = AnswerRelevancyMetric()

evaluate([test_case], [metric])
```
